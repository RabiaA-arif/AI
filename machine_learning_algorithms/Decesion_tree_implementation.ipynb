```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
# iris.data
```

```python
# iris.target
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
# independent features  and dependent features
X=df.iloc[:,:-1]
y=iris.target
```

```python
# X,y
```

```python
## train test split data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=25,random_state=10)
```

```python
X_train.head()
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# spliting dataset using decesion tree 
```

```python
treemodel= DecisionTreeClassifier(max_depth=7)
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#  prediction
Y_prediction=treemodel.predict(X_test)
```

```python
Y_prediction
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(Y_prediction,y_test)
score
```

```python
print(classification_report(Y_prediction,y_test))
```
