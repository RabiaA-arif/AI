## floating point decision in llm

```python
import torch
# # 1. Define a high-precision value (FP32/standard Python float)

original_value=20.123343434
original_torch= torch.tensor(original_value,dtype=torch.float32)
print(f"original value of {original_torch.item()}")


```

```python
#  converting FP32 into FP16(half precision)
fp16_tensor=original_torch.half()
fp16_value=fp16_tensor.item()
print(f"fp16 value {fp16_value}")
```

```python
# function that initialize the FP4 quantization
def quantize_fp4_like(val,scale_factor=4.0):
    scaled_val=val/scale_factor #1 scale the value up
    # 2 round the value
    rounded_val=round(scaled_val)
    #  3 scale the value back dwon
    fp4_like_val=rounded_val *scale_factor
    # print(f" FP4-like quantized value: {fp4_like_val}")
    return fp4_like_val
#  apply fp4 quantization to fp16
fp4_like_quantization=quantize_fp4_like(fp16_value)

print(f" fp4 simulated quantization {fp4_like_quantization}")
```

```python
# calculate error
error_fp16=abs(original_value - fp16_value)
error_fp4_like=abs(original_value - fp4_like_quantization)

print("Error comparasion")
print(f"fp16 absolute error { error_fp16}")
print(f"fp4_like absolute error {error_fp4_like}")
```

### insight
fp4 error is grater than fp16 error showing that aggressive 4 bit quantization causes a greater loss of original information(precision)
