```python
# Question 16
import time
def compare_performence(n):
    arr1 = list(range(n))
    arr2 = list(range(n))
    arr3 = list(range(n))

    # pop()
    strat = time.time()
    arr1.pop()
    t_pop = time.time() - strat

    # pop(0)
    strat = time.time()
    arr2.pop(0)
    t_pop0 = time.time() - strat


    # del arr[0]
    start = time.time()
    del arr3[0]
    t_del = time.time() - start

    return t_pop ,t_pop0, t_del
# table

# elements | pop()              | pop(0)                | del[0]
# 1,000  |(3.337860107421875e-06|1.6689300537109375e-06,|2.1457672119140625e-06
# 1,0000 |9.5367431640625e-07   |7.62939453125e-06      |7.62939453125e-06
# 1,00000|2.6226043701171875e-06|0.0001621246337890625  |0.00015044212341308594

# why is pop(o) slow

obj1 =compare_performence(1000)
# print(obj1)
obj2=compare_performence(10000)
# print(obj2)
obj3=compare_performence(100000)
print(obj3)
```

## Difference between arr.sort() vs sorted(arr)

### *return value*

arr.sort : return no value

sorted(arr): return the new list

### *original modified*

arr.sort : chnge the original value of list

sorted(arr): does not chnage the original list

### *memory*

arr.sort(): good for memory means memory efficient becaue it work on similar list

sorted(arr):it take more memory space because it create new list

### *speed*

`Use sort() if you are explicitly working with a list, do not need to preserve the original order of the items, and want to optimize your code's performance`


`Use sorted() when working with immutable types, building functional programming pipelines, or when you need to retain the original ordering of your data for later comparisons`


### *use case*

arr.sort():work only for list

sorted(arr): work on any iterable lists ,tuple,dictionary,sets

```python
# Question 18
def find_index_val(arr,val):
    indices = []
    for ind , vlue in enumerate(arr):
        if vlue == val:
            print(f"Index:{ind} of value:{val}")
            indices.append(ind)

    return {
        "Indexes":indices
    }


my_arr = [1,3,5,2,6,8,3,5,9,1,2]
num = 3
print(find_index_val(my_arr,num))

my_arr1 = [5,2,8,2,9,5,2,1]
print(find_index_val(my_arr1,2)) # [1,3,6]
print(find_index_val(my_arr1,5))# [0,5]
print(find_index_val(my_arr1,99)) # []
```
