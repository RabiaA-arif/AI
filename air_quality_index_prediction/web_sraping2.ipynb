```python
from bs4 import  BeautifulSoup  
import requests
```

```python
url="https://en.wikipedia.org/wiki/List_of_national_capitals#List"
```

```python
headers={'User-Agent': 'Mymlproject/0.0 (https://en.wikipedia.org/wiki/List_of_national_capitals#List; rabiaarifai55@gmail.com)'}
responce=requests.get(url,headers=headers)
```

```python
print(responce)
```

```python
soup=BeautifulSoup(responce.text,'html')
```

```python
soup.find_all('table')
```

```python
# table=soup.find('table',class_="wikitable sortable sticky-header jquery-tablesorter")
```

```python
table=soup.find('table',class_= "wikitable sortable sticky-header")
```

```python
table
```

```python
world_titles=table.find_all('th')
```

```python
world_titles
```

```python
# soup.find('table',class_="sidebar-heading")
```

```python
world_titles_table=[title.text.strip() for title in world_titles]
print(world_titles_table)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=world_titles_table)
df
```

```python
column_data=table.find_all('td')
```

```python
column_data
```

```python
for row in column_data[1:]:
   row_data= row.find_all('td')
   individual_row_data=[data.text.strip() for data in row_data]
   # print(individual_row_data)
   length=len(df)
   df.loc[length]=individual_row_data
```
